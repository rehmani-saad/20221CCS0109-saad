```python
import matplotlib.pyplot as plt
y=[1,2,3,4]
mylabels =[ "Apples", "Banana", "Mango", "kiwi"]

plt.pie(y, labels=mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")
plt.subplot(2,2,2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
colors=["g","r","b","y"]
plt.bar(x,y,color=colors)
plt.xlabel("X axis", fontsize=25)
plt.ylabel("Y axis", fontsize=25)
plt.title("Bar plot")
plt.show()
```

```python
x=[1,2,3,4]
y=[4,2,3,5]
colors=["g","r","b","y"]
plt.scatter(x,y,color=colors,marker="*")
plt.xlabel("X axis", fontsize=50)
plt.ylabel("Y axis", fontsize=45)
plt.title("Scatter plot")
plt.show()
```
