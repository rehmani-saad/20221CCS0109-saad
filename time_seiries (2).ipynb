```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    "Dates":pd.to_datetime(["2022-01-01","2022-01-01","2022-02-02","2022-03-03","2022-04-04","2022-05-05","2022-06-06",]),
    "Value":[10,15,22,28,35,42,48]
}

df=pd.DataFrame(data)
df.set_index("Dates",inplace=True)

plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='X')
plt.xlabel("Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01"]),
    "value":[10,20,17,24,16,18,15,20,17]
    }
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("Date")
plt.ylabel("values")
plt.title("Time series with seasonality")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text_data = "hey how you doing"
WordCloud = WordCloud(width=800, height=400, background_color='Black').generate(text_data)

plt.figure(figsize=(10,5))
plt.bar(Counter(text_data).keys(), Counter(text_data).values())
plt.title("Histogram of word frequencies")
plt.xlabel("words")
plt.ylabel("frequency")
plt.xticks(rotation=90)
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stopwords(string):
  stop_words = set(stopwords.words('english'))
  words = string.split()
  filtered_words = (word for word in words if word.lower() not in stop_words)
  return ' '.join(filtered_words)

input_string = 'this is an example sentence to remove stop words from'
result = remove_stopwords(input_string)
print("Original string:", input_string)
print("Modified string:", result)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of word
text=("Python Python Python roar grg grg hehe boo boo doo goo gee lesgooo")

# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
